```python
import tensorflow as tf
```

```python

import pandas as pd
import requests
import io
url = "https://raw.githubusercontent.com/Shubha23/Insurance-Forecast-A-Regression-Problem/master/Insurance.csv" # Make sure the url is the raw version of the file on GitHub
download = requests.get(url).content
df = pd.read_csv(io.StringIO(download.decode('utf-8')))
df
```

```python
import matplotlib.pyplot as plt
```

```python
insurance=pd.get_dummies(df)
```

```python
insurance.head()
```

```python
insurance.corr(method='pearson')
```

```python
X=insurance.drop('charges',axis=1)
y=insurance['charges']
```

```python
X.head()
```

```python
y.head()
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
X_train
```

```python
tf.random.set_seed(42)
insurance_model=tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(1)
])
insurance_model.compile(loss=tf.keras.losses.MAE,
                        optimizer=tf.keras.optimizers.Adam(learning_rate=.01),
                        metrics=['mae'])

#insurance_model.fit(X_train,y_train,epochs=100,verbose=0)
```

```python
insurance_model.evaluate(X_test,y_test)
```

***Lets Try After Normalization***

```python

new_insurance=df
new_insurance
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
ct=make_column_transformer((MinMaxScaler(),["age","bmi","children"]),
                            (OneHotEncoder(),['sex','smoker','region']))
X_x=new_insurance.drop("charges",axis=1)
y_y=new_insurance['charges']

X_x_train,X_x_test,y_y_train,y_y_test=train_test_split(X_x,y_y,test_size=.2,random_state=42)
ct.fit(X_x_train)
X_train_normal=ct.transform(X_x_train)
X_test_normal=ct.transform(X_x_test)
```

```python
X_x_train.shape, X_train_normal.shape
```

```python
X_train_normal[0]
```

```python
X_train_normal=tf.constant(X_train_normal)
X_test_normal=tf.constant(X_test_normal)
```

```python
 
insurance_model.fit(X_train_normal,y_y_train,epochs=100,verbose=0)
```

```python
insurance_model.evaluate(X_test_normal,y_y_test)
```
